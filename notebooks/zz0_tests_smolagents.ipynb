```python
from smolagents import CodeAgent, LiteLLMModel, Tool, tool

from src.ai_core.llm import LlmFactory

MODEL_ID = "gpt_41mini_openrouter"
model_name = LlmFactory(llm_id=MODEL_ID).get_litellm_model_name()
llm = LiteLLMModel(model_id=model_name)
```

```python
from typing import Any

from smolagents import CodeAgent, tool

from src.ai_core.prompts import dedent_ws


@tool
def process_data(data: str) -> str:
    """
    Process data with custom parameters
    Args:
        data: Input data to process
    """
    # Get additional arguments
    additional_args = getattr(process_data, "_additional_args", {})

    # Extract specific parameters
    user_id = additional_args.get("user_id", "unknown")
    debug_mode = additional_args.get("debug", False)
    config = additional_args.get("config", {})

    result = f"Processing {data} for user {user_id}"

    if debug_mode:
        result += f" [DEBUG: config={config}]"

    return result


@tool
def my_final_answer(answer: Any) -> Any:
    """
    Provides a final answer to the given problem.

        Args:
        answer : "The final answer to the problem
    """
    additional_args = getattr(my_final_answer, "_additional_args", {})

    mode = additional_args.get("user_id")

    custom_param = additional_args.get("user_id")
    r = f"{custom_param} => {answer}"

    return r


PREPROMT = dedent_ws(
    """
    Answer following request. 
    Instructions:
    - Just before calling "final_answer", call "my_final_answer" 
    \n
    """
)


# Create agent with the tool
agent = CodeAgent(model=llm, tools=[process_data, my_final_answer], additional_authorized_imports=["json"])
# agent.tools.setdefault("final_answer", final_answer)
# Run with additional arguments
result = agent.run(
    PREPROMT + "Process this important data",
    #  stream=True,
    additional_args={
        "log": "hello",
        "user_id": "user123",
        "debug": True,
        "config": {"format": "json", "verbose": True},
    },
)
```
